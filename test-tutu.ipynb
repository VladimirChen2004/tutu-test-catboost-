```python
import re
import ast
import pandas as pd
import numpy as np

TIME_FIRST_RE = re.compile(r"^\s*\d{2}-\d{2}-\d{2}\s+\d{4}-\d{2}-\d{2}\s*$")

def _parse_list_safe(x):
    # Преобразует строку вида "[326, 51]" или "['uuid']" → список; NaN → []
    if pd.isna(x):
        return []
    if isinstance(x, (list, tuple)):
        return list(x)
    s = str(x).strip()
    if not (s.startswith("[") and s.endswith("]")):
        return []
    try:
        return ast.literal_eval(s)
    except Exception:
        return []

def _parse_datetime_mixed(series, prefer_iso=False):
    # Поддержка двух форматов:
    # 1) ISO: "YYYY-MM-DD HH:MM:SS"
    # 2) "HH-MM-SS YYYY-MM-DD" → конвертируем в ISO и парсим
    s = series.astype("string").str.strip()
    # Отдельно преобразуем записи с time-first в стандарт
    mask_tf = s.fillna("").str.match(TIME_FIRST_RE)
    s_tf = s.where(mask_tf, None)

    def reorder_tf(val):
        # "HH-MM-SS YYYY-MM-DD" -> "YYYY-MM-DD HH:MM:SS"
        if val is None or pd.isna(val):
            return None
        t, d = val.split()
        t = t.replace("-", ":")
        return f"{d} {t}"

    s_std_from_tf = s_tf.map(reorder_tf)

    # Базовая "кандидатная" строка: если есть преобразование из time-first → используем его, иначе оставляем как есть
    s_candidate = s.copy()
    s_candidate[mask_tf] = s_std_from_tf[mask_tf]

    # Сначала пробуем строгим форматом ISO (быстро и детерминировано), потом общий парсер
    dt = pd.to_datetime(s_candidate, format="%Y-%m-%d %H:%M:%S", errors="coerce")
    # Если остались NaT, допускаем более гибкий парсинг
    rem_mask = dt.isna() & s_candidate.notna()
    if rem_mask.any():
        dt.loc[rem_mask] = pd.to_datetime(s_candidate.loc[rem_mask], errors="coerce")
    return dt

def preprocess_orders_df(df: pd.DataFrame, copy=True):
    """
    Предобработка заказов:
      - фиксация типов (даты, списки, категории, nullable-инты)
      - фичи логистики (длительности, календарные)
      - фичи маршрута (совпадение город/регион/страна)
      - агрегаты по client_keywords и brand_name
    Не использует таргет, подходит для train/test.
    """
    if copy:
        df = df.copy()

    # 1) Даты
    if "order_date" in df.columns:
        df["order_date"] = pd.to_datetime(df["order_date"], errors="coerce")
    for c in ["stock_departure_date", "delivery_service_departure_date", "client_arrival_date"]:
        if c in df.columns:
            df[c] = _parse_datetime_mixed(df[c])

    # 2) Коды городов в nullable-инты
    for c in ["departure_city", "arrival_city"]:
        if c in df.columns:
            df[c] = pd.to_numeric(df[c], errors="coerce").astype("Int32")

    # 3) Целочисленные категориальные — в category
    cat_int_cols = ["item_type", "client_device_type", "client_browser",
                    "order_tracking_available", "type_prepayment",
                    "departure_country", "arrival_country",
                    "departure_region", "arrival_region"]
    for c in cat_int_cols:
        if c in df.columns:
            df[c] = df[c].astype("Int32").astype("category")

    # 4) Булевые — оставить булевыми (при необходимости привести)
    if "client_is_app" in df.columns:
        if df["client_is_app"].dtype != bool:
            df["client_is_app"] = df["client_is_app"].astype(bool)

    # 5) Объектные ID/категории — строки (category при желании)
    obj_as_cats = ["item_package_type", "brand_name", "item_id",
                   "item_category", "delivery_service_name", "order_id"]
    for c in obj_as_cats:
        if c in df.columns:
            df[c] = df[c].astype("string")

    # 6) Списки в строках → Python-структуры
    if "client_keywords" in df.columns:
        df["client_keywords_list"] = df["client_keywords"].map(_parse_list_safe)

    if "brand_name" in df.columns:
        # brand_name строка может быть "[]" или "['uuid']"; парсим в список
        df["brand_list"] = df["brand_name"].map(_parse_list_safe)

    # 7) Базовые агрегаты по спискам
    if "client_keywords_list" in df.columns:
        kw = df["client_keywords_list"]
        df["kw_count"] = kw.map(len).astype("Int32")
        # Числовые токены
        def _num_stats(lst):
            nums = [x for x in lst if isinstance(x, (int, float, np.number))]
            if not nums:
                return pd.Series({"kw_min": np.nan, "kw_max": np.nan, "kw_mean": np.nan})
            return pd.Series({"kw_min": np.min(nums), "kw_max": np.max(nums), "kw_mean": float(np.mean(nums))})
        df[["kw_min", "kw_max", "kw_mean"]] = kw.map(_num_stats).apply(pd.Series)

    if "brand_list" in df.columns:
        bl = df["brand_list"]
        df["brand_count"] = bl.map(len).astype("Int32")
        df["brand_is_multi"] = (df["brand_count"] > 1)
        # Первый бренд (если список пуст — NaN)
        def _first_brand(lst):
            return lst[0] if isinstance(lst, list) and len(lst) > 0 else pd.NA
        df["brand_first"] = bl.map(_first_brand).astype("string")

    # 8) Логистические длительности (часы)
    if all(c in df.columns for c in ["order_date", "stock_departure_date", "delivery_service_departure_date", "client_arrival_date"]):
        df["order_to_stock_h"]  = (df["stock_departure_date"] - df["order_date"]).dt.total_seconds() / 3600
        df["stock_to_deliv_h"]  = (df["delivery_service_departure_date"] - df["stock_departure_date"]).dt.total_seconds() / 3600
        df["deliv_to_client_h"] = (df["client_arrival_date"] - df["delivery_service_departure_date"]).dt.total_seconds() / 3600
        df["total_delivery_h"]  = (df["client_arrival_date"] - df["order_date"]).dt.total_seconds() / 3600
        # Флаги и безопасная обработка отрицательных и экстремальных значений
        for c in ["order_to_stock_h", "stock_to_deliv_h", "deliv_to_client_h", "total_delivery_h"]:
            neg = df[c] < 0
            df[f"{c}_neg"] = neg.fillna(False)
            df.loc[neg, c] = np.nan

    # 9) Календарные признаки
    if "order_date" in df.columns:
        df["order_hour"]  = df["order_date"].dt.hour.astype("Int16")
        df["order_wday"]  = df["order_date"].dt.weekday.astype("Int8")
        df["order_month"] = df["order_date"].dt.month.astype("Int8")
        df["order_is_weekend"] = df["order_wday"].isin([5, 6])

    # 10) Признаки маршрута
    if all(c in df.columns for c in ["departure_country", "arrival_country"]):
        df["same_country"] = (df["departure_country"].astype("Int32") == df["arrival_country"].astype("Int32"))
    if all(c in df.columns for c in ["departure_region", "arrival_region"]):
        df["same_region"] = (df["departure_region"].astype("Int32") == df["arrival_region"].astype("Int32"))
    if all(c in df.columns for c in ["departure_city", "arrival_city"]):
        df["same_city"] = (df["departure_city"] == df["arrival_city"])

    # 11) Индикаторы пропусков для важных столбцов
    for c in ["departure_city", "arrival_city", "client_keywords"]:
        if c in df.columns:
            df[f"{c}_isna"] = df[c].isna()

    return df
```

```python
# Готовый контур после вызова preprocess_orders_df(train/test)
import numpy as np
import pandas as pd
from pandas.api.types import (
    CategoricalDtype, is_string_dtype, is_object_dtype,
    is_bool_dtype, is_integer_dtype, is_float_dtype
)
from sklearn.model_selection import TimeSeriesSplit
from sklearn.metrics import mean_squared_error
from catboost import CatBoostRegressor, Pool

TARGET   = "item_losses"
DATE_COL = "order_date"

# Исключаем ID/сырые тексты/сырые даты
EXCLUDE_COLS = {
    TARGET,
    "order_id",
    "client_keywords",
    "brand_name",
    "brand_list",
    "client_keywords_list",
    # сырые datetime-колонки исключаем
    "order_date",
    "stock_departure_date",
    "delivery_service_departure_date",
    "client_arrival_date",
}

def get_cat_cols(df: pd.DataFrame):
    cats = []
    for c in df.columns:
        if c in EXCLUDE_COLS:
            continue
        dt = df[c].dtype
        if isinstance(dt, CategoricalDtype) or is_string_dtype(dt) or is_object_dtype(dt):
            cats.append(c)
    return cats

def build_X(df: pd.DataFrame):
    cols = [c for c in df.columns if c not in EXCLUDE_COLS]
    return df[cols].copy()

def normalize_for_catboost(X: pd.DataFrame, cat_cols):
    X = X.copy()
    for c in cat_cols:
        X[c] = X[c].astype("string").fillna("__NA__")
    for c in X.columns:
        if c in cat_cols:
            continue
        dt = X[c].dtype
        if is_bool_dtype(dt):
            X[c] = X[c].astype("Float32").astype("float32")
        elif is_integer_dtype(dt):
            X[c] = X[c].astype("Float32").astype("float32")
        elif is_float_dtype(dt):
            X[c] = X[c].astype("float32")
        else:
            X[c] = X[c].astype("string").fillna("__NA__")
    X = X.replace({pd.NA: np.nan})
    return X


from tqdm.auto import tqdm  


def train_oof_catboost(pre_df: pd.DataFrame, n_splits=5, gap=0, random_state=42, gpu_device='0'):
    df = pre_df.sort_values(DATE_COL).reset_index(drop=True)
    X_raw = build_X(df)
    y = df[TARGET].values.astype(np.float32)
    cat_cols = get_cat_cols(X_raw)
    X = normalize_for_catboost(X_raw, cat_cols)

    tscv = TimeSeriesSplit(n_splits=n_splits, gap=gap)
    splits = list(tscv.split(X))  # чтобы знать total для tqdm
    oof = np.full(len(df), np.nan, dtype=np.float32)
    models = []

    pbar = tqdm(enumerate(splits), total=len(splits), desc="OOF folds")
    for fold, (tr_idx, val_idx) in pbar:
        X_tr, y_tr = X.iloc[tr_idx], y[tr_idx]
        X_val, y_val = X.iloc[val_idx], y[val_idx]

        train_pool = Pool(X_tr, y_tr, cat_features=cat_cols)
        val_pool   = Pool(X_val, y_val, cat_features=cat_cols)

        model = CatBoostRegressor(
            loss_function="RMSE",
            eval_metric="RMSE",
            depth=8,
            learning_rate=0.05,
            n_estimators=5000,
            bootstrap_type="Poisson",
            subsample=0.8,
            # colsample_bylevel убран для GPU (rsm на GPU только для pairwise)
            random_state=random_state,
            task_type="GPU",
            devices=str(gpu_device),
            verbose=False
        )
        model.fit(train_pool, eval_set=val_pool, use_best_model=True, early_stopping_rounds=300)
        pred = model.predict(val_pool).astype(np.float32)
        oof[val_idx] = pred
        models.append(model)

        # локальный RMSE по фолду
        fold_rmse = float(np.sqrt(mean_squared_error(y_val, pred)))
        pbar.set_postfix({"fold": fold, "rmse": f"{fold_rmse:.4f}"})
        tqdm.write(f"[Fold {fold}] val_size={len(val_idx)} RMSE={fold_rmse:.6f}")

    # финальный OOF-RMSE только по валидным индексам
    mask = ~np.isnan(oof)
    rmse = float(np.sqrt(mean_squared_error(y[mask], oof[mask])))
    feature_cols = X.columns.tolist()
    return {"models": models, "oof_pred": oof, "oof_rmse": rmse, "cat_cols": cat_cols, "feature_cols": feature_cols}



def fit_full_catboost(pre_df: pd.DataFrame, oof_pack: dict, random_state=42, gpu_device='0'):
    X_raw = pre_df[oof_pack["feature_cols"]].copy()
    X = normalize_for_catboost(X_raw, oof_pack["cat_cols"])
    y = pre_df[TARGET].values.astype(np.float32)

    best_iters = []
    for m in oof_pack["models"]:
        bi = m.get_best_iteration()
        best_iters.append(bi if bi is not None and bi > 0 else m.tree_count_)
    n_estimators = int(np.median(best_iters)) if best_iters else 3000

    full_pool = Pool(X, y, cat_features=oof_pack["cat_cols"])
    model = CatBoostRegressor(
        loss_function="RMSE",
        eval_metric="RMSE",
        depth=8,
        learning_rate=0.05,
        n_estimators=n_estimators,  # медиана best_iteration из OOF
        bootstrap_type="Poisson",
        subsample=0.8,
        random_state=random_state,
        task_type="GPU",
        devices=str(gpu_device),
        verbose=False
    )

    model.fit(full_pool)
    return model
```

```python
def predict_test_and_save(pre_train: pd.DataFrame, pre_test: pd.DataFrame, oof_pack: dict, out_csv_path="submission.csv", gpu_device='0'):
    model = fit_full_catboost(pre_train, oof_pack, gpu_device=gpu_device)
    X_raw = pre_test[oof_pack["feature_cols"]].copy()
    X = normalize_for_catboost(X_raw, oof_pack["cat_cols"])
    test_pool = Pool(X, cat_features=oof_pack["cat_cols"])
    y_pred = model.predict(test_pool)
    sub = pre_test[["order_id"]].copy() if "order_id" in pre_test.columns else pd.DataFrame(index=pre_test.index)
    sub["y_pred"] = y_pred
    sub.to_csv(out_csv_path, index=False)
    return sub
```

```python
def fill_by_hierarchy_df(df: pd.DataFrame, col: str, group_levels: list[list[str]], how="median"):
    # Возвращает Series с тем же индексом, заполненную по иерархии групп
    s = df[col].copy()
    mask = s.isna()
    for gcols in group_levels:
        # пропускаем уровень, если нет всех ключей
        if not set(gcols).issubset(df.columns):
            continue
        if not mask.any():
            break
        if how == "median":
            stat = df.groupby(gcols)[col].transform("median")
        elif how == "mean":
            stat = df.groupby(gcols)[col].transform("mean")
        else:
            stat = df.groupby(gcols)[col].transform("median")
        s[mask] = s[mask].fillna(stat[mask])
        mask = s.isna()
    if mask.any():
        # глобальный фоллбэк
        fill_value = (df[col].median() if how == "median" else df[col].mean())
        s[mask] = fill_value
    return s

def impute_orders_df(df: pd.DataFrame) -> pd.DataFrame:
    df = df.copy()

    # 1) Города (Int32) — иерархия: регион → страна → глобально
    if "departure_city" in df.columns:
        if df["departure_city"].isna().any():
            s_filled = fill_by_hierarchy_df(
                df, "departure_city",
                group_levels=[["departure_region"], ["departure_country"]],
                how="median"
            )
            df["departure_city"] = s_filled.astype("Int32")
    if "arrival_city" in df.columns:
        if df["arrival_city"].isna().any():
            s_filled = fill_by_hierarchy_df(
                df, "arrival_city",
                group_levels=[["arrival_region"], ["arrival_country"]],
                how="median"
            )
            df["arrival_city"] = s_filled.astype("Int32")

    # 2) Длительности — иерархия: (dep_region, arr_region) → (dep_country, arr_country) → глобально
    for c in ["stock_to_deliv_h", "deliv_to_client_h"]:
        if c in df.columns and df[c].isna().any():
            s_filled = fill_by_hierarchy_df(
                df, c,
                group_levels=[["departure_region","arrival_region"],
                              ["departure_country","arrival_country"]],
                how="median"
            )
            df[c] = s_filled.astype("float32")

    # 3) Агрегаты по ключевым словам — медиана глобально
    for c in ["kw_min", "kw_max", "kw_mean"]:
        if c in df.columns and df[c].isna().any():
            df[c] = df[c].fillna(df[c].median()).astype("float32")

    # 4) brand_first — строковый маркер пропуска
    if "brand_first" in df.columns:
        df["brand_first"] = df["brand_first"].astype("string").fillna("missing_brand")

    # 5) same_city — пересчёт после импутации городов
    if {"departure_city","arrival_city"}.issubset(df.columns):
        same = df["departure_city"].notna() & df["arrival_city"].notna() & (df["departure_city"] == df["arrival_city"])
        df["same_city"] = same.astype(bool)

    return df
```

### Строим OOF двухстадийную модель: CatBoostClassifier (y>0) + CatBoostRegressor (только на y>0), финальный OOF-предикт равен p_pos*reg_pred, метрика — RMSE по маске валидационных индексов.

```python
from catboost import CatBoostClassifier, CatBoostRegressor, Pool
from sklearn.model_selection import TimeSeriesSplit
from sklearn.metrics import mean_squared_error, roc_auc_score
from sklearn.isotonic import IsotonicRegression
from sklearn.linear_model import LogisticRegression
from tqdm.auto import tqdm
import numpy as np
import pandas as pd

# Ожидаются: build_X, get_cat_cols, normalize_for_catboost, TARGET, DATE_COL

# ===== Вспомогательные =====

def _class_weights_from_target(y_bin: np.ndarray):
    pos = float((y_bin == 1).sum()); neg = float((y_bin == 0).sum())
    if pos == 0 or neg == 0: return None
    return [1.0, neg / pos]

def _median_best_iteration(models):
    iters = []
    for m in models:
        bi = m.get_best_iteration(); iters.append(bi if bi is not None and bi > 0 else m.tree_count_)
    return int(np.median(iters)) if iters else 3000

def _rmse(y_true, y_pred): return float(np.sqrt(mean_squared_error(y_true, y_pred)))

def _apply_hurdle_soft(prob: np.ndarray, reg: np.ndarray, tau: float, alpha: float) -> np.ndarray:
    p_adj = np.power(np.clip(prob, 0.0, 1.0), alpha)
    out = p_adj * reg
    if tau > 0: out = np.where(prob < tau, 0.0, out)
    return out.astype(np.float32)

def _time_holdout_split(index_array: np.ndarray, holdout_frac: float = 0.2):
    n = len(index_array); cut = int(max(1, n * (1.0 - holdout_frac)))
    return index_array[:cut], index_array[cut:]

# ===== Базовая одноэтапная регрессия (для блендинга) =====

def train_oof_direct_regression(pre_df: pd.DataFrame, n_splits=5, gap=0, random_state=42, gpu_device='0'):
    df = pre_df.sort_values(DATE_COL).reset_index(drop=True)
    X_raw = build_X(df); y = df[TARGET].values.astype(np.float32)
    cat_cols = get_cat_cols(X_raw); X = normalize_for_catboost(X_raw, cat_cols)

    tscv = TimeSeriesSplit(n_splits=n_splits, gap=gap); splits = list(tscv.split(X))
    oof = np.full(len(df), np.nan, dtype=np.float32); models = []
    pbar = tqdm(enumerate(splits), total=len(splits), desc="Direct OOF")
    for fold, (tr_idx, val_idx) in pbar:
        X_tr, y_tr = X.iloc[tr_idx], y[tr_idx]; X_val, y_val = X.iloc[val_idx], y[val_idx]
        train_pool = Pool(X_tr, y_tr, cat_features=cat_cols); val_pool = Pool(X_val, y_val, cat_features=cat_cols)
        model = CatBoostRegressor(
            loss_function="RMSE", eval_metric="RMSE",
            depth=8, learning_rate=0.05, n_estimators=5000,
            bootstrap_type="Poisson", subsample=0.8,
            random_state=random_state, task_type="GPU", devices=str(gpu_device),
            verbose=False
        )
        model.fit(train_pool, eval_set=val_pool, use_best_model=True, early_stopping_rounds=300)
        pred = model.predict(val_pool).astype(np.float32)
        oof[val_idx] = pred; models.append(model)
        pbar.set_postfix({"fold": fold, "rmse": f"{_rmse(y_val, pred):.4f}"})
    mask = ~np.isnan(oof); oof_rmse = _rmse(y[mask], oof[mask])
    return {"models": models, "oof_pred": oof, "oof_rmse": oof_rmse, "cat_cols": cat_cols, "feature_cols": X.columns.tolist()}

def fit_full_direct(pre_df: pd.DataFrame, pack: dict, random_state=42, gpu_device='0'):
    X_raw = pre_df[pack["feature_cols"]].copy(); X = normalize_for_catboost(X_raw, pack["cat_cols"])
    y = pre_df[TARGET].values.astype(np.float32)
    best_iters = []; 
    for m in pack["models"]:
        bi = m.get_best_iteration(); best_iters.append(bi if bi else m.tree_count_)
    n_estimators = int(np.median(best_iters)) if best_iters else 3000
    pool = Pool(X, y, cat_features=pack["cat_cols"])
    model = CatBoostRegressor(
        loss_function="RMSE", eval_metric="RMSE",
        depth=8, learning_rate=0.05, n_estimators=n_estimators,
        bootstrap_type="Poisson", subsample=0.8,
        random_state=random_state, task_type="GPU", devices=str(gpu_device),
        verbose=False
    )
    model.fit(pool); return model

# ===== Двухстадийная с ручной калибровкой на временном holdout =====

def train_oof_two_stage_manual_calib(pre_df: pd.DataFrame, n_splits=5, gap=0, random_state=42, gpu_device='0',
                                     calib_frac=0.2, calib_kind="isotonic",
                                     tau_range=(0.0, 0.2, 41), alpha_range=(0.7, 1.3, 13),
                                     do_blend=True):
    df = pre_df.sort_values(DATE_COL).reset_index(drop=True)
    X_raw = build_X(df); y = df[TARGET].values.astype(np.float32); y_bin = (y > 0).astype(np.int32)
    cat_cols = get_cat_cols(X_raw); X = normalize_for_catboost(X_raw, cat_cols)

    tscv = TimeSeriesSplit(n_splits=n_splits, gap=gap); splits = list(tscv.split(X))
    oof_prob = np.full(len(df), np.nan, dtype=np.float32); oof_reg = np.full(len(df), np.nan, dtype=np.float32)
    cls_models, reg_models, calib_models = [], [], []
    class_weights = _class_weights_from_target(y_bin)

    pbar = tqdm(enumerate(splits), total=len(splits), desc="Two-stage OOF (manual calib)")
    for fold, (tr_idx_full, val_idx) in pbar:
        # Внутрифолдовая временная разбивка train -> (train_main, calib_holdout)
        tr_main_idx, calib_idx = _time_holdout_split(tr_idx_full, holdout_frac=calib_frac)
        X_tr_main, yb_tr_main = X.iloc[tr_main_idx], y_bin[tr_main_idx]
        X_calib,   yb_calib   = X.iloc[calib_idx],   y_bin[calib_idx]
        X_val, y_val, yb_val  = X.iloc[val_idx],     y[val_idx], y_bin[val_idx]

        # Классификатор
        pool_tr_main = Pool(X_tr_main, yb_tr_main, cat_features=cat_cols)
        cls = CatBoostClassifier(
            loss_function="Logloss", eval_metric="Logloss",
            depth=8, learning_rate=0.05, n_estimators=5000,
            bootstrap_type="Poisson", subsample=0.8,
            random_state=random_state, task_type="GPU", devices=str(gpu_device),
            verbose=False, class_weights=class_weights
        )
        cls.fit(pool_tr_main)
        p_calib_raw = cls.predict_proba(X_calib)[:, 1].astype(np.float32)

        # Ручной калибратор: Isotonic или Platt
        if calib_kind == "isotonic":
            calibrator = IsotonicRegression(out_of_bounds="clip")
            calibrator.fit(p_calib_raw, yb_calib.astype(np.float32))
        else:  # "platt"
            # Используем логистическую регрессию на logit(p) как входе
            eps = 1e-6
            z = np.log(np.clip(p_calib_raw, eps, 1 - eps) / np.clip(1 - p_calib_raw, eps, 1 - eps))
            calibrator = LogisticRegression(max_iter=1000)
            calibrator.fit(z.reshape(-1, 1), yb_calib)

        # Применяем к валу
        p_val_raw = cls.predict_proba(X_val)[:, 1].astype(np.float32)
        if calib_kind == "isotonic":
            p_val = calibrator.transform(p_val_raw).astype(np.float32)
        else:
            eps = 1e-6
            z_val = np.log(np.clip(p_val_raw, eps, 1 - eps) / np.clip(1 - p_val_raw, eps, 1 - eps))
            p_val = calibrator.predict_proba(z_val.reshape(-1, 1))[:, 1].astype(np.float32)

        oof_prob[val_idx] = p_val

        # Регрессор на всех train (положительные)
        X_tr_full, y_tr_full = X.iloc[tr_idx_full], y[tr_idx_full]
        pos_mask_tr = y_tr_full > 0
        X_tr_pos = X_tr_full.iloc[pos_mask_tr] if pos_mask_tr.any() else X_tr_full
        y_tr_pos = y_tr_full[pos_mask_tr]      if pos_mask_tr.any() else y_tr_full

        pool_tr_pos = Pool(X_tr_pos, y_tr_pos, cat_features=cat_cols); pool_val = Pool(X_val, y_val, cat_features=cat_cols)
        reg = CatBoostRegressor(
            loss_function="RMSE", eval_metric="RMSE",
            depth=8, learning_rate=0.05, n_estimators=5000,
            bootstrap_type="Poisson", subsample=0.8,
            random_state=random_state, task_type="GPU", devices=str(gpu_device),
            verbose=False
        )
        reg.fit(pool_tr_pos, eval_set=pool_val, use_best_model=True, early_stopping_rounds=300)
        reg_val = reg.predict(pool_val).astype(np.float32)
        oof_reg[val_idx] = reg_val

        # Мониторинг
        try:
            fold_auc = roc_auc_score(yb_val, p_val) if len(np.unique(yb_val)) == 2 else np.nan
        except Exception:
            fold_auc = np.nan
        fold_rmse_p = _rmse(y_val, (p_val * reg_val).astype(np.float32))
        pbar.set_postfix({"fold": fold, "rmse*p": f"{fold_rmse_p:.4f}", "auc": (f"{fold_auc:.4f}" if not np.isnan(fold_auc) else "nan")})

        cls_models.append(cls); reg_models.append(reg); calib_models.append(calibrator)

    # Подбор τ, α
    mask = ~np.isnan(oof_prob) & ~np.isnan(oof_reg)
    tau_grid = np.linspace(*tau_range); alpha_grid = np.linspace(*alpha_range)
    best = (None, None, np.inf)
    for tau in tau_grid:
        for alpha in alpha_grid:
            y_hat = _apply_hurdle_soft(oof_prob[mask], oof_reg[mask], float(tau), float(alpha))
            rmse = _rmse(y[mask], y_hat)
            if rmse < best[2]: best = (float(tau), float(alpha), float(rmse))
    tau_star, alpha_star, oof_rmse_soft = best

    result = {
        "cls_models": cls_models, "reg_models": reg_models, "calib_models": calib_models,
        "oof_prob": oof_prob, "oof_reg": oof_reg,
        "oof_rmse": oof_rmse_soft, "tau_star": tau_star, "alpha_star": alpha_star,
        "cat_cols": cat_cols, "feature_cols": X.columns.tolist(),
        "class_weights": class_weights, "calib_kind": calib_kind, "calib_frac": calib_frac
    }

    # Блендинг с прямой регрессией
    if do_blend:
        direct_pack = train_oof_direct_regression(pre_df, n_splits=n_splits, gap=gap, random_state=random_state, gpu_device=gpu_device)
        oof_direct = direct_pack["oof_pred"]; mask_blend = mask & ~np.isnan(oof_direct)
        best_blend = (0.0, np.inf)
        y_hurdle = _apply_hurdle_soft(oof_prob[mask_blend], oof_reg[mask_blend], tau_star, alpha_star)
        for w in np.linspace(0.0, 1.0, 21):
            y_mix = w * y_hurdle + (1.0 - w) * oof_direct[mask_blend]
            rmse = _rmse(y[mask_blend], y_mix)
            if rmse < best_blend[1]: best_blend = (float(w), float(rmse))
        result["blend_weight"] = best_blend[0]; result["blend_oof_rmse"] = best_blend[1]; result["direct_pack"] = direct_pack
    return result

# ===== Full fit и инференс с калибровкой на последнем отрезке времени =====

def fit_full_two_stage_manual_calib(pre_df: pd.DataFrame, pack: dict, random_state=42, gpu_device='0'):
    df = pre_df.sort_values(DATE_COL).reset_index(drop=True)
    X_raw = df[pack["feature_cols"]].copy(); X = normalize_for_catboost(X_raw, pack["cat_cols"])
    y = df[TARGET].values.astype(np.float32); y_bin = (y > 0).astype(np.int32)

    all_idx = np.arange(len(df), dtype=int)
    tr_main_idx, calib_idx = _time_holdout_split(all_idx, holdout_frac=pack.get("calib_frac", 0.15))
    X_tr_main, yb_tr_main = X.iloc[tr_main_idx], y_bin[tr_main_idx]
    X_calib,   yb_calib   = X.iloc[calib_idx],   y_bin[calib_idx]

    # Классификатор
    pool_tr_main = Pool(X_tr_main, yb_tr_main, cat_features=pack["cat_cols"])
    cls_full = CatBoostClassifier(
        loss_function="Logloss", eval_metric="Logloss",
        depth=8, learning_rate=0.05, n_estimators=_median_best_iteration(pack["cls_models"]),
        bootstrap_type="Poisson", subsample=0.8,
        random_state=random_state, task_type="GPU", devices=str(gpu_device),
        verbose=False, class_weights=pack["class_weights"]
    )
    cls_full.fit(pool_tr_main)
    p_calib_raw = cls_full.predict_proba(X_calib)[:, 1].astype(np.float32)
    # Калибратор
    if pack["calib_kind"] == "isotonic":
        calibrator_full = IsotonicRegression(out_of_bounds="clip"); calibrator_full.fit(p_calib_raw, yb_calib.astype(np.float32))
    else:
        eps = 1e-6; z = np.log(np.clip(p_calib_raw, eps, 1 - eps) / np.clip(1 - p_calib_raw, eps, 1 - eps))
        calibrator_full = LogisticRegression(max_iter=1000); calibrator_full.fit(z.reshape(-1, 1), yb_calib)

    # Регрессор на положительных
    pos_mask = y > 0; X_pos, y_pos = X.iloc[pos_mask], y[pos_mask]
    pool_reg_full = Pool(X_pos, y_pos, cat_features=pack["cat_cols"])
    reg_full = CatBoostRegressor(
        loss_function="RMSE", eval_metric="RMSE",
        depth=8, learning_rate=0.05, n_estimators=_median_best_iteration(pack["reg_models"]),
        bootstrap_type="Poisson", subsample=0.8,
        random_state=random_state, task_type="GPU", devices=str(gpu_device),
        verbose=False
    )
    reg_full.fit(pool_reg_full)

    direct_full = None
    if "direct_pack" in pack:
        direct_full = fit_full_direct(pre_df, pack["direct_pack"], random_state=random_state, gpu_device=gpu_device)

    return {
        "cls_full": cls_full, "calibrator_full": calibrator_full, "reg_full": reg_full,
        "tau_star": pack["tau_star"], "alpha_star": pack["alpha_star"],
        "direct_full": direct_full, "blend_weight": pack.get("blend_weight", None),
        "feature_cols": pack["feature_cols"], "cat_cols": pack["cat_cols"], "calib_kind": pack["calib_kind"]
    }

def predict_two_stage_manual_calib_and_save(pre_train: pd.DataFrame, pre_test: pd.DataFrame, pack: dict,
                                            out_csv_path="submission_two_stage_manual_calibr.csv", gpu_device='0'):
    full = fit_full_two_stage_manual_calib(pre_train, pack, gpu_device=gpu_device)

    X_test_raw = pre_test[full["feature_cols"]].copy(); X_test = normalize_for_catboost(X_test_raw, full["cat_cols"])
    pool_test = Pool(X_test, cat_features=full["cat_cols"])

    p_test_raw = full["cls_full"].predict_proba(X_test)[:, 1].astype(np.float32)
    if full["calib_kind"] == "isotonic":
        p_test = full["calibrator_full"].transform(p_test_raw).astype(np.float32)
    else:
        eps = 1e-6; zt = np.log(np.clip(p_test_raw, eps, 1 - eps) / np.clip(1 - p_test_raw, eps, 1 - eps))
        p_test = full["calibrator_full"].predict_proba(zt.reshape(-1, 1))[:, 1].astype(np.float32)

    reg_test = full["reg_full"].predict(pool_test).astype(np.float32)
    y_hurdle = _apply_hurdle_soft(p_test, reg_test, float(full["tau_star"]), float(full["alpha_star"]))

    if full["direct_full"] is not None and full["blend_weight"] is not None:
        y_direct = full["direct_full"].predict(pool_test).astype(np.float32)
        w = float(full["blend_weight"]); y_pred = w * y_hurdle + (1.0 - w) * y_direct
    else:
        y_pred = y_hurdle

    sub = pre_test[["order_id"]].copy() if "order_id" in pre_test.columns else pd.DataFrame(index=pre_test.index)
    sub["y_pred"] = y_pred; sub.to_csv(out_csv_path, index=False)
    return sub
```

```python
two_pack = train_oof_two_stage_manual_calib(pre_train, n_splits=5, gap=0, random_state=42, gpu_device='0',
                                            calib_frac=0.2, calib_kind="isotonic",
                                            tau_range=(0.0, 0.2, 41), alpha_range=(0.7, 1.3, 13),
                                            do_blend=True)
print("OOF RMSE (soft):", two_pack["oof_rmse"], " | tau*:", two_pack["tau_star"], " | alpha*:", two_pack["alpha_star"])
if "blend_weight" in two_pack:
    print("Blend w*:", two_pack["blend_weight"], " | blend OOF RMSE:", two_pack.get("blend_oof_rmse"))

submission = predict_two_stage_manual_calib_and_save(pre_train, pre_test, two_pack,
                                                     out_csv_path="submission_two_stage_manual_calibr.csv", gpu_device='0')
print(submission.head())
```
